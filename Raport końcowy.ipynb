# Import modułów i danych

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import analiza_eksploracyjna as ae
from scipy import stats as st
from scipy.stats import mannwhitneyu as mw
import seaborn as sns
import missingno as msno

dane = pd.read_csv('dataset.csv', index_col=0)
pozytywy = dane[dane.label == 'positive']
negatywy = dane[dane.label == 'negative']
```

# Liczebność danych

```python
ilosc_obs = dane.shape[0]; ilosc_poz = pozytywy.shape[0]; ilosc_neg = negatywy.shape[0]
ilosc_none = ilosc_obs - ilosc_poz - ilosc_neg
plt.figure(figsize = (15,15))
plt.title('Procentowa ilość obserwacje dla wartości zmiennej \'label\'', fontsize = 20)
lab = ['Pozytywne', 'Negatywne', 'Braki']
ae.kolowy([ilosc_poz, ilosc_neg, ilosc_none], lab, ['red', 'green', 'blue'])

print('Ilość obserwacji w całym zbiorze: ', ilosc_obs)
print('Ilość zmiennych: ', dane.shape[1])
print('\nIlość obserwacji dla przypadków pozytywnych: ', ilosc_poz)
print('Ilość obserwacji dla przypadków negatywnych: ', ilosc_neg)
print('Ilość obserwacji dla braku diagnozy: ', ilosc_none)
```

# Korelacje Spearmana

```python
korelacje = dane.corr(method = 'spearman')
pary_skorelowane = ae.korelacje(korelacje)
```

```python
ae.skladowe(pary_skorelowane)
```

Grafy korelacji zmiennych

```python
normalne_grupa1 = ['x2', 'x5', 'x21', 'x23', 'x24', 'x25', 'x31', 'x33', 'x34', 'x43']
normalne_grupa2 = ['x7', 'x29', 'x42', 'x46', 'x47']
plt.figure(figsize = (20,8))
plt.subplot(1, 2, 1)
ae.grafy_korelacji(pary_skorelowane, normalne_grupa1); 
plt.subplot(1, 2, 2)
ae.grafy_korelacji(pary_skorelowane, normalne_grupa2)
```
